```python
import pandas as pd
import numpy as np
```

```python
#file generated by gan which contains interest and satisfaction values
df=pd.read_csv('/output_valueszero.csv')
df.head()
```

```python
#df = pd.read_csv('/content/output_values_updated.csv')

# Identify non-integer values in the 'userId' column
non_integer_values = df[~df['userId'].astype(str).str.isdigit()]['userId']

# Replace non-integer values with NaN
df.loc[~df['userId'].astype(str).str.isdigit(), 'userId'] = np.nan

# Convert the 'userId' column to float and then to integers
df['userId'] = df['userId'].astype(float).astype('Int32')
print(df)
```

```python
#df = pd.read_csv('/content/output_values_updated.csv')

# Identify non-integer values in the 'userId' column
non_integer_values = df[~df['movieId'].astype(str).str.isdigit()]['movieId']

# Replace non-integer values with NaN
df.loc[~df['userId'].astype(str).str.isdigit(), 'movieId'] = np.nan

# Convert the 'userId' column to float and then to integers
df['movieId'] = df['movieId'].astype(float).astype('Int32')
print(df)
```

```python



df['interest'] = pd.to_numeric(df['interest'], errors='coerce')

# Clean non-numeric values in the "satisfaction" column
df['satisfaction'] = pd.to_numeric(df['satisfaction'], errors='coerce')

# Drop rows with NaN values resulting from non-numeric conversions
df.dropna(subset=['interest', 'satisfaction'], inplace=True)

# Calculate the difference between "interest" and "satisfaction"
df['serendipity_score'] = df['satisfaction'] - df['interest']


# Identify non-integer values in the 'userId' column
#non_integer_values = df[~df['interest'].astype(str).str.isdigit()]['interest']
#non_integer_values = df[~df['satisfaction'].astype(str).str.isdigit()]['satisfaction']
non_integer_values = df[~df['userId'].astype(str).str.isdigit()]['userId']
non_integer_values = df[~df['movieId'].astype(str).str.isdigit()]['movieId']

# Replace non-integer values with NaN
#df.loc[~df['interest'].astype(str).str.isdigit(), 'interest'] = np.nan
#df.loc[~df['satisfaction'].astype(str).str.isdigit(), 'satisfaction'] = np.nan
df.loc[~df['userId'].astype(str).str.isdigit(), 'userId'] = np.nan
df.loc[~df['movieId'].astype(str).str.isdigit(), 'movieId'] = np.nan

# Convert the 'userId' column to float and then to integers

df['userId'] = df['userId'].astype(float).astype('float32')
df['movieId'] = df['movieId'].astype(float).astype('float32')


df['userId'] += 1

# Save the modified DataFrame back to a CSV file

df.head()
```

```python
import pandas as pd

# Assuming your DataFrame is named 'df'
# Group by 'userId' and 'movieId' and calculate the minimum values of 'interest' and 'satisfaction'
grouped_df = df.groupby(['userId', 'movieId']).agg({'interest': 'min', 'satisfaction': 'min'}).reset_index()

# Create a new DataFrame to store the results
df = pd.DataFrame({
    'userId': grouped_df['userId'],
    'movieId': grouped_df['movieId'],
    'min_interest': grouped_df['interest'],
    'min_satisfaction': grouped_df['satisfaction']
})

# Save the resulting DataFrame to a CSV file
#result_df.to_csv('/content/drive/MyDrive/phase 2/min_values.csv', index=False)
```
